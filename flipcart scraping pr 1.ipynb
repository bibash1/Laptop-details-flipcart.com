```python
from selenium import webdriver
from selenium.webdriver import Chrome
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time 
import pandas as pd

s=Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=s)
driver.maximize_window()
```

```python
max=30000
min=45000
```

```python
driver.get("https://www.flipkart.com/search?q=laptop&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&p%5B%5D=facets.price_range.from%3D"+str(min)+"&p%5B%5D=facets.price_range.to%3D"+str(max))
```

```python
nextpage_xpath="//a[@class='_1LKTO3']"
price_xpath="//div[@class='_30jeq3 _1_WHN1']"

   
```

```python

product_title=[]
product_price=[]
product_rating=[]
 
```

```python
for i in range(43):
    title_xpath="//div[@class='_4rR01T']"
    
    
    title=driver.find_elements_by_xpath(title_xpath)
    price=WebDriverWait(driver, 10).until(
       EC.presence_of_all_elements_located((By.XPATH, price_xpath))
    )
    master_div=driver.find_elements_by_xpath("//div[@class='col col-7-12']")
    for j in master_div:
        try:
            product_rating.append(j.find_element_by_xpath(".//div[@class='gUuXy-']").text)
  
        except:
            product_rating.append("NO_rating_given")
   
       
        
    
    nextpage= WebDriverWait(driver, 50).until(
       EC.presence_of_element_located((By.XPATH, nextpage_xpath))
    )   
    for each_title in title:
        product_title.append(each_title.text)
    for each_price in price:
        x=each_price.text.replace("â‚¹","")
       
        product_price.append(x)
       
                         
    nextpage.click()
    time.sleep(3)
driver.close()   
    
    
```
